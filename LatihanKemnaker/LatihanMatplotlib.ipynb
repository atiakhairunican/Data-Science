```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Data Contoh')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'b*-')
```

```python
fig = plt.figure()

# 0.1,0.1 adalah jarak x,y dari titik 0,0
# 0.8,0.8 adalah panjang dan lebar dari plot
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y,'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Contoh Judul')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('X utama')
axes1.set_ylabel('Y utama')
axes1.set_title('Plot Utama')

axes2.plot(x,y,'r')
axes2.set_xlabel('X dalam')
axes2.set_ylabel('Y dalam')
axes2.set_title('Plot Dalam')
```

```python
# (8,4) adalah besar kanvas sebesar 800x400
# semakin besar dpi maka semakin tajam gambar
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(5,4), dpi=100)

axes.plot(x,y,'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Plot Judul')

# Jika dpi sudah dideklarasikan sebelumnya maka tidak perlu dideklarasikan lagi
fig.savefig("contoh.png", dpi=100)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label="Data x pangkat 2")
axes.plot(x,x**3,label="Data x pangkat 3")

axes.legend(loc=0)
```

Style Plot

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x+1,color="red",linewidth=0.25)
axes.plot(x,x+2,color="red",linewidth=0.50)
axes.plot(x,x+3,color="red",linewidth=0.75)
axes.plot(x,x+4,color="red",linewidth=1)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x+1,color="green",lw=3,linestyle='-')
axes.plot(x,x+2,color="green",lw=3,ls='-.')
axes.plot(x,x+3,color="green",lw=3,ls=':')
axes.plot(x,x+4,color="green",lw=3,ls='--')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

line, = axes.plot(x,x+1,color="black",lw=1.50)
# [5,5,15,5] adalah panjang garis yg pendek, jarak antar garis,
# panjang garis putus yang panjang, jarak garis putus yang pendek dengan yang panjang
line.set_dashes([5,5,15,5])
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

# Simbol - simbol apa saja yang dapat digunakan dapat dilihat di matplotlib.org/3.1.1/api/markers_api.html
axes.plot(x,x+1,color="blue",lw=3,ls='-',marker='+')
axes.plot(x,x+2,color="blue",lw=3,ls='-',marker='o')
axes.plot(x,x+3,color="blue",lw=3,ls='-',marker='s')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x+1,color="blue",lw=3,ls='-',marker='+',markersize='10')
axes.plot(x,x+2,color="blue",lw=3,ls='-',marker='o',markerfacecolor='red')
axes.plot(x,x+3,color="blue",lw=3,ls='-',marker='s',markerfacecolor='yellow',markeredgewidth=3,markeredgecolor="green")
```
